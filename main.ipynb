Data Cleaning

```python
all_games.isnull().sum()
```

```python
all_games['summary'] = all_games['summary'].fillna("No Summary")
```

```python
all_games.isnull().sum()
```

```python
all_games = all_games.drop(columns=['index'])
```

```python
all_games.head(5)
```

```python
all_games['release_date'] = pd.to_datetime( all_games['release_date'])
```

```python
all_games.info()
```

```python
# errors = 'coerce' to handle values like tbd
all_games['user_review'] = pd.to_numeric(
    all_games['user_review'],
    errors='coerce'
)
```

```python
all_games.info()
```

```python
all_games['clean_title'] = (
    all_games['name']
    .str.lower()
    .str.strip()
    .str.replace(r'[^a-z0-9\s]', '', regex=True)
    .str.replace(r'\s+', ' ', regex=True)
)
```

```python
video_games['clean_title'] = (
    video_games['Title']
    .str.lower()
    .str.strip()
    .str.replace(r'[^a-z0-9\s]', '', regex=True)
    .str.replace(r'\s+', ' ', regex=True)
)
```

```python
all_games['platform_clean'] = (
    all_games['platform']
    .str.lower()
    .str.strip()
)
```

```python
video_games['platform_clean'] = (
    video_games['Release.Console']
    .str.lower()
    .str.strip()
)
```

```python
sorted(all_games['platform_clean'].unique())
```

```python
sorted(video_games['platform_clean'].unique())
```

```python
platform_map = {
    'nintendo ds': 'ds',
    'nintendo wii': 'wii',
    'sony psp': 'psp',
    'x360': 'xbox 360'
}
```

```python
video_games['platform_clean'] = (
    video_games['platform_clean']
    .replace(platform_map)
)
```

```python
sorted(video_games['platform_clean'].unique())
```

```python
merged = pd.merge(
    all_games,
    video_games,
    on=['clean_title', 'platform_clean'],
    how='inner'
)
```

```python
merged = merged.drop(columns=['clean_title'])
```

```python
df = merged[[
    'name',
    'platform',
    'release_date',
    'meta_score',
    'user_review',
    'Metrics.Sales',
    'Metadata.Genres',
    'Release.Year'
]].copy()
```

```python
# rename some columms to much simpler names
df = df.rename(columns={
    'Metrics.Sales': 'sales',
    'Metadata.Genres': 'genre',
    'Release.Year': 'year'
})
```

Visualization

```python
top10_games = df.sort_values(by='sales', ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(data=top10_games, x='sales', y='name')

plt.title("Top 10 Games by Sales")
plt.xlabel("Sales")
plt.ylabel("Game")
plt.show()
```

```python
# Sales per platform
platform_sales = df.groupby('platform')['sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(12,6))
plt.bar(platform_sales.index, platform_sales.values)

plt.title("Total Sales per Platform")
plt.xlabel("Platform")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
# Trend of sales
sales_trend = df.groupby('year')['sales'].sum()

plt.figure(figsize=(12,6))
plt.plot(sales_trend.index, sales_trend.values, marker='o')

plt.title("Sales Trend Over Years")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.grid(True)
plt.show()
```

```python
# Sales over platform
plt.figure(figsize=(14,6))

sns.boxplot(x='platform', y='sales', data=df)

plt.title("Sales Distribution by Platform")
plt.xlabel("Platform")
plt.ylabel("Sales")
plt.xticks(rotation=45)

plt.show()
```

```python
# Relation between numeric features
numeric_df = df.select_dtypes(include=['int64', 'float64'])
corr = numeric_df.corr()

plt.figure(figsize=(8,6))

sns.heatmap(corr, annot=True, cmap='coolwarm')

plt.title("Correlation Between Numeric Features")
plt.show()
```
